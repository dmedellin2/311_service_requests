```python
labels = ['1st','2nd','3rd']
color = ['royalblue', 'indianred']

nypd_complaints = [sr_df[sr_df['complaint_type']=='Noise - Residential']['waittime'].mean(),
                  sr_df[sr_df['complaint_type']=='Illegal Parking']['waittime'].mean(),
                  sr_df[sr_df['complaint_type']=='Blocked Driveway']['waittime'].mean()]

hpd_complaints = [sr_df[sr_df['complaint_type']=='HEAT/HOT WATER']['waittime'].mean(),
                  sr_df[sr_df['complaint_type']=='UNSANITARY CONDITION']['waittime'].mean(),
                  sr_df[sr_df['complaint_type']=='PLUMBING']['waittime'].mean()]


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, nypd_complaints, width, label='NYPD',color=color[0])
rects2 = ax.bar(x + width/2, hpd_complaints, width, label="HPD",color=color[1])

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Most Frequent Complaint Type',size = 12)
ax.set_ylabel('Average Wait Time (Days)',size = 12)
ax.set_title('Average Wait Time for Most Frequent\nComplaint Type per Agency',size=16)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()



fig.tight_layout()

plt.show()
```
